## Bar plot

```python
# plotting barplot

res = sns.barplot(data=mtcars, x='cyl', y='mpg')
plt.show()
```

```python
res = sns.barplot(x = 'cyl', y = 'mpg', data=mtcars, color='y' )
plt.show()
```

```python
res = sns.barplot(x='am', y='mpg', data=mtcars, palette='rocket')
plt.show()
```

```python
res = sns.barplot(x='gear', y='hp', data=mtcars, palette='Set1')
plt.show()
```

### Scatter Plot

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.species.value_counts()
```

```python
# relation between sepal length and petal length

sns.scatterplot(x='sepal_length', y='petal_length', data=iris)
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', data=iris, hue= 'species')
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', data=iris, hue='species', size='species')
```

### Pairplot

```python
sns.pairplot(iris)
```

### Lmplot

```python
sns.lmplot(x='petal_length', y='petal_width', data=iris)
```

```python
sns.lmplot(x='petal_length', y='petal_width', data=iris, hue='species', markers=['o', '*', '^'])
```
