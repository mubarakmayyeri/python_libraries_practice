```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.randint(10, size=(10,))
y = np.random.randint(200, size=(10,))
```

### Scatter plot

```python
plt.scatter(x, y, c='g', marker='d')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Plot')
```

```python
x = np.arange(1, 21)
y = x*x

plt.plot(x, y, 'r*:' )
```
